<h2>Función suma en Naturales</h2 >

````python
def Suma(n):
    if n == 0:
        return n
    else:
        return n + Suma(n-1)
````

<h2>Funcion de las potencias de 2</h2>

````python
def pot2(n) : 
    if n == 0: 
        return 1 
    else : 
        return 2* pot2(n-1)
````

<h2> ¿Qué es un arbol binario en python? </h2>

En python, se definen los arboles binarios sobre la programacion orientada a objetos, Los "Arboles" son una clase en python que creamos nosotros, y de ahi en adelante todos los arboles que creemos son objetos como nosotros querramos crear el arbol en cuestion. La clase tree se muestra a continuacion:

````python
class Tree:
    def __init__(self, left, right):
        self.left = left
        self.right = right
        pass
````

En los arboles binarios de Python, la estructura base son las denominadas hojas, que son los nodos que no tienen hijos. A partir de estas hojas, se construyen los nodos internos, que pueden tener uno o dos hijos. La clase `Tree` define un nodo con dos atributos: `left` y `right`, que representan los hijos izquierdo y derecho del nodo, respectivamente.

<h2> ¿Cómo se define un arbol binario en python? </h2>

```python
hoja = Tree(None, None)  
Arbol1 = Tree(hoja, hoja)
Arbol2 = Tree(hoja, Arbol1)
```
Aca vemos como se crea un arbol binario en python, primero creamos una hoja, que es un nodo sin hijos, luego creamos un arbol con dos hojas, y luego creamos otro arbol con una hoja y el arbol que acabamos de crear. Es importante tener en cuenta que los arboles que creamos tambien pueden ser hojas, por lo que podemos crear arboles de cualquier tamaño y forma, siempre y cuando sigamos la estructura de la clase `Tree`.


````python
h = Tree(None, None)
r1 = Tree(h, h)
r2 = Tree(h, r1)
k1 = Tree(h, h)
k2 = Tree(k1, k1)
t1 = Tree(h, h)
t2 = Tree(h, t1)
t3 = Tree(h, t2)

#Estos son los arboles que hay en el taller #1
````

<h2> Función que cuenta hojas de un árbol </h2>

````python
def numHojas(A):
    if A.left is None:
        return 1
    else : 
        return 0 + numHojas(A.left) + numHojas(A.right)
````

<h2> Función que cuenta las Aristas de un árbol </h2>

````python
def Num_Aristas(A):
    if A.left is None:
        return 0
    else: 
        return 2 + Num_Aristas(A.left) + Num_Aristas(A.right)
````

<h2> Función que cuenta los nodos en un árbol </h2>

````python
def Num_nodos(A):
    if A.left is None:
        return 1
    else :
        return 1 + Num_nodos(A.left) + Num_nodos(A.right)
````

<h2> Demostracion de la funcion Num_Aristas </h2>

Como es una proposicion en arboles binarios, esta se puede demostrar mediante induccion estructural, que es una forma de demostrar proposiciones en estructuras recursivas como los arboles. La induccion estructural se basa en dos pasos:
<lu>
    <li>Paso base: Se demuestra que la proposicion es verdadera para el caso base, que en este caso es la hoja del arbol.</li>
    <li>Paso inductivo: Se demuestra que si la proposicion es verdadera para un los arboles que salen del arbol A (Hijo izquierdo e Hijo dereho) </li>
</lu>

Entonces, la proposicion es verdadera para el arbol A. 

Bajo este esquema, podemos empezar a demostrar la funcion `Num_Aristas`
```python


Proposicion: num_Aristas(h) es par para todo arbol binario.

Demostracion:

Paso base: 
Sea h una hoja, observe que num_Aristas(h) = 0 y 0 es par, por lo tanto la proposicion es verdadera cuando el arbol es una hoja.

Paso inductivo: 
Sea A un arbol de la forma A = Tree(B, C), suponga por H.I que num_Aristas(B) y num_Aristas(C) son pares, ed, 2|num_Aristas(B) y 2|num_Aristas(C), ed existen n,m pertenecientes a los enteros tq Num_aristas(B) = 2n y Num_aristas(C) = 2m. Observe que num_aristas(A) = 2 + num_aristas(B) + num_aristas(C); num_aristas(A) = 2 + 2n + 2m = 2(1+n+m); luego 2|num_aristas(A), por lo tanto, num_aristas(A) es par.

Por induccion estructural, la proposicion es verdadera para todo arbol binario.
```
